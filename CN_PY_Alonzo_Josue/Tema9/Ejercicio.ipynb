```python
import os

class Netflix:
    def __init__(self, nombre, descripcion, director, categoria):
        self.nombre = nombre
        self.descripcion = descripcion
        self.director = director
        self.categoria = categoria
        
        directorio_actual = os.path.expanduser("~/Documents/CN_PY_Alonzo_Josue/Tema9")
        
        self.directorio_base = os.path.join(directorio_actual, "peliculas")
        self.ruta_categoria = os.path.join(self.directorio_base, categoria)
        os.makedirs(self.ruta_categoria, exist_ok=True)

    def guardar_datos(self):
        archivo_categoria = os.path.join(self.ruta_categoria, f"{self.nombre}.txt")
        with open(archivo_categoria, "w", encoding="utf-8") as f:
            f.write(f"Nombre: {self.nombre}\n")
            f.write(f"Descripción: {self.descripcion}\n")
            f.write(f"Director: {self.director}\n")
            f.write(f"Categoría: {self.categoria}\n")

class Spotify:
    def __init__(self, nombre, descripcion, compositor, categoria):
        self.nombre = nombre
        self.descripcion = descripcion
        self.compositor = compositor
        self.categoria = categoria
        
        directorio_actual = os.path.expanduser("~/Documents/CN_PY_Alonzo_Josue/Tema9")
        
        self.directorio_base = os.path.join(directorio_actual, "canciones")
        self.ruta_categoria = os.path.join(self.directorio_base, categoria)
        os.makedirs(self.ruta_categoria, exist_ok=True)

    def guardar_datos(self):
        archivo_categoria = os.path.join(self.ruta_categoria, f"{self.nombre}.txt")
        with open(archivo_categoria, "w", encoding="utf-8") as f:
            f.write(f"Nombre: {self.nombre}\n")
            f.write(f"Descripción: {self.descripcion}\n")
            f.write(f"Compositor: {self.compositor}\n")
            f.write(f"Categoría: {self.categoria}\n")

# Lista de películas
peliculas = [
    ("Inception", "Un ladrón que roba secretos a través de los sueños.", "Christopher Nolan", "Ciencia Ficción"),
    ("Titanic", "Un romance épico en el trágico hundimiento del Titanic.", "James Cameron", "Drama"),
    ("El Padrino", "Historia de una familia mafiosa italiana.", "Francis Ford Coppola", "Crimen"),
    ("Interstellar", "Exploración del espacio y agujeros de gusano.", "Christopher Nolan", "Ciencia Ficción"),
    ("Gladiador", "Un general romano en busca de venganza.", "Ridley Scott", "Acción"),
    ("The Dark Knight", "Batman contra el Joker en Gotham.", "Christopher Nolan", "Acción"),
    ("Pulp Fiction", "Varias historias criminales interconectadas.", "Quentin Tarantino", "Crimen"),
    ("La La Land", "Una historia de amor musical en Hollywood.", "Damien Chazelle", "Romance"),
    ("Avatar", "Una historia en el planeta Pandora.", "James Cameron", "Fantasía"),
    ("The Matrix", "Un hacker descubre la realidad virtual.", "Lana y Lilly Wachowski", "Ciencia Ficción")
]

# Lista de canciones
canciones = [
    ("Bohemian Rhapsody", "Una obra maestra del rock de Queen.", "Freddie Mercury", "Rock"),
    ("Imagine", "Un himno de paz y esperanza.", "John Lennon", "Pop"),
    ("Stairway to Heaven", "Una de las mejores canciones de rock de todos los tiempos.", "Led Zeppelin", "Rock"),
    ("Billie Jean", "Un éxito mundial de Michael Jackson.", "Michael Jackson", "Pop"),
    ("Smells Like Teen Spirit", "El himno del grunge de Nirvana.", "Kurt Cobain", "Rock"),
    ("Hotel California", "Una historia misteriosa en forma de canción.", "Eagles", "Rock"),
    ("Hey Jude", "Un clásico de los Beatles.", "Paul McCartney", "Pop"),
    ("Like a Rolling Stone", "Una canción legendaria de Bob Dylan.", "Bob Dylan", "Rock"),
    ("Sweet Child O' Mine", "Un éxito de Guns N' Roses.", "Axl Rose", "Rock"),
    ("Wonderwall", "Una de las canciones más icónicas de Oasis.", "Noel Gallagher", "Rock"),
    ("Shape of You", "Un éxito de Ed Sheeran.", "Ed Sheeran", "Pop"),
    ("Havana", "Una canción latina de Camila Cabello.", "Camila Cabello", "Latina"),
    ("Lose Yourself", "El himno de la superación personal de Eminem.", "Eminem", "Hip Hop"),
    ("Rolling in the Deep", "Una potente canción de Adele.", "Adele", "Soul"),
    ("Uptown Funk", "Un éxito funk de Mark Ronson y Bruno Mars.", "Mark Ronson & Bruno Mars", "Funk"),
    ("Despacito", "Un hit global de reggaetón de Luis Fonsi.", "Luis Fonsi", "Reggaetón"),
    ("Bad Guy", "Un éxito pop de Billie Eilish.", "Billie Eilish", "Pop"),
    ("Gangnam Style", "Un fenómeno viral de PSY.", "PSY", "K-Pop"),
    ("Thunderstruck", "Una de las canciones más famosas de AC/DC.", "AC/DC", "Rock"),
    ("Take Five", "Una famosa pieza de jazz de Dave Brubeck.", "Dave Brubeck", "Jazz")
]

# Guardar películas
for pelicula in peliculas:
    p = Netflix(*pelicula)
    p.guardar_datos()

# Guardar canciones
for cancion in canciones:
    c = Spotify(*cancion)
    c.guardar_datos()

print("Archivos creados exitosamente.")
```
