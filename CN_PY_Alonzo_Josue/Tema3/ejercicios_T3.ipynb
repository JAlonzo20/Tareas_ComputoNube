```python
# 1. Convertir a minúsculas
frutas = ["MANZANA", "BANANA", "CEREZA"]
nueva_lista = [fruta.lower() for fruta in frutas]
print(nueva_lista)

# 2. Capitalizar la primera letra
frutas = ["manzana", "banana", "cereza"]
nueva_lista = [fruta.capitalize() for fruta in frutas]
print(nueva_lista)

# 3. Contar la cantidad de letras 'a' en cada fruta
frutas = ["manzana", "banana", "cereza"]
nueva_lista = [fruta.count('a') for fruta in frutas]
print(nueva_lista)

# 4. Reemplazar letras
frutas = ["manzana", "banana", "cereza"]
nueva_lista = [fruta.replace('a', '@') for fruta in frutas]
print(nueva_lista)

# 5. Verificar si la fruta empieza con 'b'
frutas = ["manzana", "banana", "cereza"]
nueva_lista = [fruta.startswith('b') for fruta in frutas]
print(nueva_lista)

# 6. Verificar si la fruta termina con 'a'
frutas = ["manzana", "banana", "cereza"]
nueva_lista = [fruta.endswith('a') for fruta in frutas]
print(nueva_lista)

# 7. Obtener la longitud de cada palabra
frutas = ["manzana", "banana", "cereza"]
nueva_lista = [len(fruta) for fruta in frutas]
print(nueva_lista)

# 8. Invertir cada palabra
frutas = ["manzana", "banana", "cereza"]
nueva_lista = [fruta[::-1] for fruta in frutas]
print(nueva_lista)

# 9. Reemplazar espacios (aunque no hay, es demostrativo)
frutas = ["manzana", "banana", "cereza"]
nueva_lista = [fruta.replace(" ", "_") for fruta in frutas]
print(nueva_lista)

# 10. Quitar la primera y última letra
frutas = ["manzana", "banana", "cereza"]
nueva_lista = [fruta[1:-1] for fruta in frutas]
print(nueva_lista)

# 11. Convertir a título (cada palabra con mayúscula inicial)
frutas = ["manzana roja", "banana amarilla", "cereza dulce"]
nueva_lista = [fruta.title() for fruta in frutas]
print(nueva_lista)

# 12. Verificar si todas las letras son alfabéticas
frutas = ["manzana", "banana", "cereza", "pera123"]
nueva_lista = [fruta.isalpha() for fruta in frutas]
print(nueva_lista)

# 13. Llenar con ceros hasta 10 caracteres
frutas = ["manzana", "banana", "cereza"]
nueva_lista = [fruta.zfill(10) for fruta in frutas]
print(nueva_lista)

# 14. Separar cada palabra en listas de letras
frutas = ["manzana", "banana", "cereza"]
nueva_lista = [list(fruta) for fruta in frutas]
print(nueva_lista)

# 15. Eliminar espacios en blanco (strip)
frutas = ["  manzana  ", " banana", "cereza "]
nueva_lista = [fruta.strip() for fruta in frutas]
print(nueva_lista)
```
